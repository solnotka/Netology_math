```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.uniform()
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
rv_exp = sts.expon()
sample = rv_exp.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv_exp.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
n10 = 10
exp_5 = 5
exp_8 = 8


samples_count = 1000
# Массивы для сохранения средних по выборкам
a1 = np.array([])
a10 = np.array([])
ae5 = np.array([])
ae8 = np.array([])

for number in range(samples_count):
    # На каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1)
    sample10 = rv.rvs(size = n10)
    samplee5 = rv.rvs(size = exp_5)
    samplee8 = rv.rvs(size = exp_8)
    # Считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    a10 = np.append(a10, sample10.mean())
    ae5 = np.append(ae5, samplee5.mean())
    ae8 = np.append(ae8, samplee8.mean())
```

```python
# Строим графики
plt.figure(figsize=(15, 6))
# Подписи для графиков
label_1 = 'sample of ' + str(n1)
label_10 = 'sample of ' + str(n10)
label_e5 = 'sample of ' + str(exp_5)
label_e8 = 'sample of ' + str(exp_8)

plt.subplot(1, 4, 1)
plt.hist(a1, density=True)
plt.title('uniform, ' + label_1)

plt.subplot(1, 4, 2)
plt.hist(a10, density=True)
plt.title('uniform, ' + label_10)

plt.subplot(1, 4, 3)
plt.hist(ae5, density=True)
plt.title('exponential, ' + label_e5)

plt.subplot(1, 4, 4)
plt.hist(ae8, density=True)
plt.title('exponential, ' + label_e8)

plt.ylabel('samples means')
plt.xlabel('$x$')
```
