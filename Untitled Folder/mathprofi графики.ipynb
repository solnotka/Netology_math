```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.e
```

```python
x = list(range(-10, 1)) + [1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9] +  list(range(2, 10))
y = [pow(np.e, 1 / (1 - a)) for a in x]
```

```python
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-100, 101)]
y = [(9 + 6 * a - 3 * pow(a, 2)) / (pow(a, 2) - 2 * a + 13) for a in x]
```

```python
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-10, 12)]
x.remove(0)
y = [(3 * pow(a, 4) + 1) / pow(a, 3) for a in x]
y
```

```python
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-100, -20)] + [a * .1 for a in range(20, 100)]
x.remove(2)
y = [np.log(pow(a, 2) - 4) for a in x]
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-100, 100)]
y = [a * pow(np.e, a * -1) for a in x]
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-100, 100)]
x.remove(0)
y = [1 / pow(a, 2) for a in x]
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-100, 100)]
y = [pow(a, 3) for a in x]
plt.plot(x, y)
```

```python
x = [a * .1 for a in range(-40, 40)]
y = [a / (pow(a, 2) + 4) for a in x]
plt.plot(x, y)
```
