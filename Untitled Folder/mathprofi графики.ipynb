```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection="3d")

x, y = np.meshgrid(np.linspace(-5, 5, 1000), np.linspace(-5, 5, 1000))
z = y ** 2 // (y - x)

ax.plot_surface(x, y, z)

plt.show()
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection="3d")

x, y = np.meshgrid(np.linspace(-5, 5, 1000), np.linspace(-5, 5, 1000))

z = x ** 2 // (2 * x + y)

ax.plot_surface(x, y, z)

plt.show()
```
