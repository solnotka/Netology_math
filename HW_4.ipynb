### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
from sympy import *
import numpy as np
import scipy.misc as misc
import matplotlib.pyplot as plt
```

```python
# Получаю значения производных с помощью sympy
x = symbols('x')

y = pow(x, 5) + 4 * sin(2 * x) + cos(3 * x + 3)

y_i = y.diff()

y_ii = y_i.diff()

print(f'Значение первой производной в точке x = 1: {y_i.evalf(subs={x: 1})}')
print(f'Значение второй производной в точке x = 1: {y_ii.evalf(subs={x: 1})}')
```

```python
# Первая производная
y_i
```

```python
# Вторая производная
y_ii
```

```python
# Получаю значения производных с помощью scipy
def f(x):
    return pow(x, 5) + 4 * sin(2 * x) + cos(3 * x + 3)

print(f'Значение первой производной в точке x = 1: {misc.derivative(f, 1.0, dx=1e-6, n=1)}')
print(f'Значение второй производной в точке x = 1: {misc.derivative(f, 1.0, dx=1e-6, n=2)}')
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
# Сначала посмотрю, что там за производные
f = pow(sin(2 * x + 1), 5)
# Первая производная
f_i = f.diff()
f_i
```

```python
# Вторая производная
f_ii = f.diff().diff()
f_ii
```

```python
# Получаю значения x и y для графиков
def f(x):
    return pow(sin(2 * x + 1), 5)

x = np.linspace(-5, 5)

y = [f(a) for a in x]

y_1 = [misc.derivative(f, a, dx = 1e-6, n = 1) for a in x]

y_2 = [misc.derivative(f, a, dx = 1e-6, n = 2) for a in x]
```

```python
# Cобственно графики
plt.figure(figsize = (5, 10))
plt.plot(x, y, label = 'function')
plt.plot(x, y_1, label = 'derivative_1')
plt.plot(x, y_2, label = 'derivative_2')
plt.legend()
plt.grid()
plt.show()
```
